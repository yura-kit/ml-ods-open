Видно, что kMeans не очень хорошо отличает только активности друг от друга. Используйте метод локтя, чтобы выбрать оптимальное количество кластеров. Параметры алгоритма и данные используем те же, что раньше, меняем только `n_clusters`.

```python
from scipy.spatial.distance import cdist

# Ваш код здесь
inertia = []
for k in tqdm_notebook(range(1, n_classes + 1)):
    kmeanModel = KMeans(n_clusters=k, random_state=RANDOM_STATE, n_init=100)
    kmeanModel.fit(X_pca)
    inertia.append(sum(np.min(cdist(X_pca, kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / X_pca.shape[0])
 
    #
```

```python
plt.plot(range(1, n_classes + 1), inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

Настройте для `LinearSVC` гиперпараметр `C` с помощью `GridSearchCV`. 

- Обучите новый `StandardScaler` на обучающей выборке (со всеми исходными признаками), примените масштабирование к тестовой выборке

```python
# Ваш код здесь
#

scaller = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
svc = LinearSVC(random_state=RANDOM_STATE)
svc_params = {'C': [0.001, 0.01, 0.1, 1, 10]}
```

```python
%%time
# Ваш код здесь

clf = GridSearchCV(svc, svc_params, cv=3)
clf.fit(X_train_scaled, y_train)
best_svc = clf.best_estimator_
```

```python
from sklearn.metrics import accuracy_score, roc_auc_score, precision_score, recall_score



preds = best_svc.predict(X_test_scaled)
# print('Accuracy: {:.5f}'.format(precision_score(y_test,  preds, average = None)))
# print('ROC: {:.5f}'.format(recall_score(y_test,  preds, average='micro')))
```

```python
# Ваш код здесь
clf.best_params_
y_predicted = best_svc.predict(X_test_scaled)
print('Accuracy: {:.5f}'.format(accuracy_score(y_test,  y_predicted)))
```
